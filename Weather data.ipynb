```python
import os
rootdir = './Weather_DataSet'

for subdir, dirs, files in os.walk(rootdir):
    #Iterate through all files in the set root directory
    for file in files:
        #choose all the files that start with 'produkt_klima_tag'
        if file.startswith('produkt_klima_tag'): 
            #create the readable path for all those files
            path = (os.path.join(subdir, file).replace("\\","/"))
            #read the respective .txt file into a datafram data1
            data1 = pd.read_csv(path, sep=';', encoding='ISO-8859-1')
            #subset to the years 2016&2017
            data1 = data1[(data1['MESS_DATUM']>=20160101)&(data1['MESS_DATUM']<=20171231)]
            #create one big dataframe
            data_all = data_all.append(data1)
```
